## _Figure 4C:_ El Capitan TOC curve.
* * *
Compiled TOC data from El Capitan Beach, with a moving average through the data depicted as a solid line, with 1σ  analytical uncertainty envelope. 
* * *

```python
ElCapTOC=plt.figure(figsize=(4,15)) #define figure dimensions
    
    
#import TOC data for El Cap
TOC_ECNew=pd.read_csv('CSV/El_Cap_Data/ElCap_TOC_New.csv') #new data from El Capitan
TOC_John=pd.read_csv('CSV/El_Cap_Data/TOC_ElCap_John2002.csv') #data extracted from John et al., 2002, integrated into our section heights
Combo_TOC=pd.read_csv('CSV/El_Cap_Data/ElCap_ComboToc.csv') #combined TOC for El Capitan
#########

#plot TOC data
plt.scatter(TOC_John['TOC'],TOC_John['Height'],marker='o',color='darkgreen', label='John et al. (2001)',zorder=6)
plt.scatter(TOC_ECNew['OrgC']*100,TOC_ECNew['SAMPLE_HEIGHT'], marker='o', color='olive', label='Jones, this study',zorder=5)
#########

#calculate moving average of El Cap TOC
LSMA_TOC =(Combo_TOC['El_Cap_TOC']*100).ewm(alpha=.05,adjust=False).mean()#exponential moving average of data
LSMA_TOCSD=(Combo_TOC['El_Cap_TOC']*100).ewm(alpha=.05,adjust=False).std()#calculate standard deviation of moving average
#########

#Resize moving average of TOC to match sed rate data.
from scipy import interpolate
f = interpolate.interp1d(Combo_TOC['El_Cap_Height'],LSMA_TOC,fill_value="extrapolate")
xnew = np.arange(0, 240, 1) #create interpolated curve of same length as sedrate data
resize_LSMA_TOC = f(xnew) #use interpolation function returned by `interp1d`
#########

#plot TOC moving average
plt.plot(resize_LSMA_TOC,xnew,color='seagreen',zorder=4)
LSMA_TOCup=resize_LSMA_TOC+1.8 #add 1 sigma analytical uncertainty envelope to moving average
LSMA_TOCdn=resize_LSMA_TOC-1.8 #add 1-sigma analytical uncertainty envelope to moving average
plt.fill_betweenx(range(0,len(xnew)),LSMA_TOCdn,LSMA_TOCup, alpha=.45, color='seagreen',zorder=4) #plot 1-sigma uncertainty envelope
#########

#plot Figure 2 strat details
elcapashes()
stratzones()
#########


#labeling params etc.
plt.legend(fontsize=10,framealpha=1)
plt.xlabel('Total Organic Carbon \n(wt%)',fontsize=16)
axes = plt.gca()
axes.set_xlim([0,25])
axes.set_ylim([0,220])
axes.tick_params(labelleft=False, length=0)
axes.tick_params(axis='x', labelsize=16 )
plt.grid(True)


#save plot
plt.savefig("Output/ElCapTOC_Figure4C.pdf")
```
